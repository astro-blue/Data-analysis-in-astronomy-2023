```python
import matplotlib.pyplot as plt
import numpy as np
import sklearn.cluster as sc
from sklearn.datasets import make_blobs
#import my_plot_style as mps
import sklearn.metrics as sm

#mps.my_plot_style()
```

### Reference https://realpython.com/k-means-clustering-python/

```python
features, true_labels = make_blobs(
    n_samples=500,
    centers=3,
    cluster_std=2.,
    random_state=100
)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(features[:,0],features[:,1])
```

```python
plt.figure(figsize=(10,8))
kmean = sc.KMeans(n_clusters=5,init='random')
kmean.fit(features)
plt.scatter(features[:,0],features[:,1],c=kmean.labels_)
plt.plot(kmean.cluster_centers_[:,0],kmean.cluster_centers_[:,1],ls='',marker='s',color='r',markersize=10)
plt.xlabel('x')
plt.ylabel('y')
```

```python
sse = []
for n in range(2,11):
    kmean = sc.KMeans(n_clusters=n,init='random')
    kmean.fit(features)
    sse.append(kmean.inertia_)
```

```python
plt.plot(np.arange(2,11),sse,ls='--',lw=3)
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
```
