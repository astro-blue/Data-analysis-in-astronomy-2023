### TAB or 4 spaces

```python
def calculate_the_volume_of_sphere(radius):
    volume = 4./3*np.pi*(radius**3)
    surface = np.pi*(radius**2)
    return {'a':volume,'b':surface}
    #return volume, surface
```

```python
r = 5
output = calculate_the_volume_of_sphere(r)

print(output['a'])
print(output['b'])

#print(volume,surface)
```

### Maplotlib

https://matplotlib.org/stable/tutorials/introductory/customizing.html

```python
x = np.arange(0,10)
y = np.arange(0,10)
```

```python
x,y
```

```python
def my_plot_style():
   
    params = {'legend.fontsize': 20,
             'axes.labelsize': 20,
             'axes.titlesize':20,
             'xtick.labelsize':20,
             'ytick.labelsize':20,
             'xtick.major.size':5,
              'xtick.minor.size':2.5,
             'ytick.major.size':5,
              'ytick.minor.size':2.5,
             'figure.facecolor':'w',
             #'lines.linewidth' : 1.5,
              'xtick.major.width':1.5,
              'ytick.major.width':1.5,
              'xtick.minor.width':1.5,
              'ytick.minor.width':1.5,
              'axes.linewidth':1.5,
              'xtick.direction':'in',
              'ytick.direction':'in',
             'ytick.labelleft':True,
              'text.usetex' : False,
             'font.family': 'sans-serif'}
  
    plt.rcParams.update(params)
```

```python
def removing_ytick_label():
    ax = plt.gca()
    ax.axes.yaxis.set_ticklabels([])
```

```python
my_plot_style()
plt.figure(figsize=(10,10))
noise = np.random.normal(0,0.5,len(x))
#plt.plot(x,y,lw=5,ls=':')
plt.errorbar(x,y+noise,yerr=0.5,xerr=0.5,ls='',color='C0',marker='s',markersize=10,capsize=8,lw=3,mew=3,label='Lan et al. (2020)',zorder=1)
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
```

```python
%matplotlib inline

noise = np.random.normal(0,0.5,len(x))

#my_plot_style()
plt.figure(figsize=(14,12))

plt.subplots_adjust(left=0.07,right=0.95,top=0.92,bottom=0.09,wspace=0.0,hspace=0.0)
plt.subplot(2,2,1)
plt.plot(x,y,label='YYYY',color='C0',zorder=0,ls='--',marker='o')
plt.errorbar(x,y+noise,yerr=0.5,xerr=0.5,ls='',color='C1',marker='o',markersize=13,capsize=8,lw=3,mew=3,label='data',zorder=1)
plt.legend(frameon=False,loc=4)

plt.xlabel('x',fontsize=20)
plt.ylabel('y',fontsize=20)
plt.title('My plot')
plt.xlim(-1,10.3)
plt.ylim(-1,10.3)


plt.subplot(2,2,2)
plt.plot(x,y,label='Best fit',color='C0',zorder=0)
plt.errorbar(x,y+noise,yerr=0.5,ls='',color='C1',marker='o',markersize=13,capsize=8,lw=3,mew=3,label='data',zorder=1)
#plt.legend(frameon=False)
plt.xlabel('x',fontsize=20)
#plt.ylabel('y',fontsize=20)
plt.title('My plot')
plt.xlim(-1,10.3)
plt.ylim(-1,10.3)


#Using the 

removing_ytick_label()

plt.subplot(2,2,3)
plt.plot(x,y,label='Best fit',color='C0',zorder=0)
plt.errorbar(x,y+noise,yerr=0.5,ls='',color='C1',marker='o',markersize=13,capsize=8,lw=3,mew=3,label='data',zorder=1)
#plt.legend(frameon=False)
plt.xlabel('x',fontsize=20)
plt.ylabel('y',fontsize=20)
#plt.title('My plot')
plt.xlim(-1,10.3)
plt.ylim(-1,10.3)

plt.text(6,3,'$\Omega_{\\rm test}$',fontsize=20)
plt.text(6,2,'$\Omega_{test}$',fontsize=20)
plt.text(6,0,'$\sum_{i=0}^\infty x_i$',fontsize=15)
plt.subplot(2,2,4)
plt.plot(x,y,label='Best fit',color='C0',zorder=0)
plt.errorbar(x,y+noise,yerr=0.5,ls='',color='C1',marker='o',markersize=13,capsize=8,lw=3,mew=3,label='data',zorder=1)
#plt.legend(frameon=False)
plt.xlabel('x',fontsize=20)
#plt.ylabel('y',fontsize=20)
#plt.title('My plot')
plt.xlim(-1,10.3)
plt.ylim(-1,10.3)
removing_ytick_label()
plt.savefig('plot_v1.pdf')
#plt.show()
```
